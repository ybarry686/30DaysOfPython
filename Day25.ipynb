```python
import pandas as pd
import numpy as np 

nums = [1, 2, 3, 4, 5]

# Creating Pandas Series with Default Index; Index --> Value
s = pd.Series(nums)
print(s)

# Creating Pandas Series with Custom Index
s = pd.Series(nums, index=[1,2,3,4,5])
print(s)

fruits = ['Orange', 'Banana', 'Mango']
fruits = pd.Series(fruits, index=[1,2,3])
print(fruits)
```

```python
# Creating Pandas Series from a Dictionary
dct = {'name':'Yagouba','country':'Guinea','city':'Conakry'}
s = pd.Series(dct) # Dict key becomes series index
print(s)

# Creating a Constant Pandas Series
s = pd.Series(10, index=[1,2,3])
print(s)

# Creating a Pandas Series Using Linspace: Give me N points between A and B
s = pd.Series(np.linspace(5, 20, 10)) # linspace(starting, end, items)
print(s)
```

```python
# Creating DataFrames from List of Lists
data = [
    ['Asabeneh', 'Finland', 'Helsink'], 
    ['David', 'UK', 'London'],
    ['John', 'Sweden', 'Stockholm']
]
df = pd.DataFrame(data, columns=['Names', 'Country', 'City'])
print(df)

# Creating DataFrames Using Dictionary
data = {'Name': ['Asabeneh', 'David', 'John'], 'Country':[
    'Finland', 'UK', 'Sweden'], 'City': ['Helsiki', 'London', 'Stockholm']}
df = pd.DataFrame(data) # Dict keys become column names
print('\n', df)

# Creating DataFrames Using Dictionary
data = [
    {'Name': 'Asabeneh', 'Country': 'Finland', 'City': 'Helsinki'},
    {'Name': 'David', 'Country': 'UK', 'City': 'London'},
    {'Name': 'John', 'Country': 'Sweden', 'City': 'Stockholm'}]
df = pd.DataFrame(data) # Each Unique dict key become a column name
print('\n', df)
```

```python
"""
Modifying a DataFrame:
    1. Create a new DataFrame
    2. Create a new column and add it to a DataFrame
    3. Remove an existing column from a DataFrame
    4. Modify an existing column in a DataFrame
    5. Change the data type of column values in the DataFrame
"""

data = [
    {"Name": "Asabeneh", "Country":"Finland","City":"Helsinki"},
    {"Name": "David", "Country":"UK","City":"London"},
    {"Name": "John", "Country":"Sweden","City":"Stockholm"}]
df = pd.DataFrame(data)
print(df)
```

```python
# Adding a column: Like adding a key to a dict
weights = [74, 78, 69]
df['Weight'] = weights
print('\n', df)

heights = [173, 175, 169]
df['Height'] = heights
print('\n', df)

# Adding BMI: Formula = Weight/(Height * Height) (height in meters) 
    # Change height from centimeters to meters
df['Height'] = df['Height'] * 0.01
print('\n', df)

def calculate_bmi():
    weights = df['Weight']
    heights = df['Height']
    bmi = []

    for w, h in zip(weights, heights):
        bmi.append(w / (h * h))
        
    return bmi

bmi = calculate_bmi()
df['BMI'] = bmi
print('\n', df)
```

```python
# Formating DataFrame Columns
df['BMI'] = round(df['BMI'], 1)
print(df)

# Adding more data
birth_year = ['1769', '1985', '1990']
current_year = pd.Series(2025, index=[0,1,2])
df['Birth Year'] = birth_year
df['Current Year'] = current_year
print('\n', df)
```

```python
# Checking Data Types of Column Values
print(df.Weight.dtype)
df['Birth Year'].dtype # Gives a str object, we should change it to an int

df['Birth Year'] = df['Birth Year'].astype('int')
print(df['Birth Year'].dtype) 

df['Current Year'] = df['Current Year'].astype('int')
print(df['Current Year'].dtype) 

ages = df['Current Year'] - df['Birth Year']
print(ages)

df['Ages'] = ages
print(df)

mean = (40 + 35) / 2
print('Mean: ', mean)
```

```python
# Boolean Indexing: Only return values that meet certain criteria
print(df[df['Ages'] >  120])
print(df[df['Ages'] <  120])
```
